```python
%matplotlib inline
import numpy as np
import qiskit
from qiskit import *

from qiskit.tools.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor

from qiskit import IBMQ
IBMQ.load_account()
sim = Aer.get_backend('qasm_simulator')
vector_sim = Aer.get_backend('statevector_simulator')
provider = IBMQ.get_provider('ibm-q'); qcomp = provider.get_backend('ibmqx2')

print(qiskit.__qiskit_version__)
```

```python
# This algorithms solves one of the many problems where quantum computers are able to outperform classical computers.
# Task is as follows:
# You have n bits that go into a blackbox, that performs either a constant operation (output only 0s or only 1s) or a balanced operation (based on the input, apply an operation, so the output is not fixed)
# The task is to find out whether the operation in the blackbox is constant or balanced. 
# Remember that a constant operation means that the black box will randomly choose 0 or 1, and it is constant in that the input doesn't make a difference
# For a classical computer, this takes 2^(n - 1) + 1 tries, so if there is one input bit, then it takes 2 tries.
# A quantum computer can do this in one try flat, and this is what this algorithm does.
# We will be using 3 input bits for this example.
n = 3
```

```python
#Creating the constant blackbox

# 1 extra qubit to write the answer to
const_oracle = QuantumCircuit(n+1)

# Randomly choose whether output should be 0 or 1
output = np.random.randint(2)
if output == 1:
    const_oracle.x(n)  #Index starts from 0 so n is last qubit

const_oracle.draw('mpl')
```

```python
#Creating the Balanced blackbox

balanced_oracle = QuantumCircuit(n+1)

#Input string 
b_str = "101"

# Flip values wherever input is 1, so as to get desired input
for qubit in range(len(b_str)):
    if b_str[qubit] == '1':
        balanced_oracle.x(qubit)

# Use barrier for some clarity
balanced_oracle.barrier()

# Controlled-NOT gates to the output bit - this way input to blackbox has effect on output
for qubit in range(n):
    balanced_oracle.cx(qubit, n)

balanced_oracle.barrier()

# Place X-gates to return all qubits to original state
for qubit in range(len(b_str)):
    if b_str[qubit] == '1':
        balanced_oracle.x(qubit)

# Show balanced blackbox
balanced_oracle.draw('mpl')
```

```python
#Now that we have made the blackboxes, we will now try and make the algorithm to solve the task:

dj_circuit = QuantumCircuit(n+1, n)

#For this algorithm, we will put all input bits into superposition in the |+> state, and the output qubit in superposition in the |-> state

for qubit in range(n):
    dj_circuit.h(qubit)

dj_circuit.x(n)
dj_circuit.h(n)

# Add the blackbox we want
dj_circuit += balanced_oracle

# Remove qubits from superposition by applying hadamard (quantum gates are reversible.)
for qubit in range(n):
    dj_circuit.h(qubit)
dj_circuit.barrier()

# Measure standard qubits
for i in range(n):
    dj_circuit.measure(i,i)

# Display circuit
dj_circuit.draw('mpl')
```

```python
#Run the circuit on simulator. Since we used balanced oracle, we expect output to be 111 consistently
results = execute(dj_circuit, backend=sim, shots=1024).result()
answer = results.get_counts()

plot_histogram(answer)
```

```python
#We can make the circuit with the constant blackbox as well, and see the result.

dj_circuit = QuantumCircuit(n+1, n)

#For this algorithm, we will put all input bits into superposition in the |+> state, and the output qubit in superposition in the |-> state

for qubit in range(n):
    dj_circuit.h(qubit)

dj_circuit.x(n)
dj_circuit.h(n)

# Add the blackbox we want --------- Line changed to constant orcale
dj_circuit += const_oracle

# Remove qubits from superposition by applying hadamard (quantum gates are reversible.)
for qubit in range(n):
    dj_circuit.h(qubit)
dj_circuit.barrier()

# Measure standard qubits
for i in range(n):
    dj_circuit.measure(i,i)

# Display circuit
dj_circuit.draw('mpl')
```

```python
#Run the circuit on simulator. Since we used constant oracle, we expect output to be 000 consistently
results = execute(dj_circuit, backend=sim, shots=1024).result()
answer = results.get_counts()

plot_histogram(answer)
```

```python
#Making circuit with balanced blackbox

dj_balanced = QuantumCircuit(n+1, n)

for qubit in range(n):
    dj_balanced.h(qubit)

dj_balanced.x(n)
dj_balanced.h(n)

# Add the blackbox we want
dj_balanced += balanced_oracle

for qubit in range(n):
    dj_balanced.h(qubit)
dj_balanced.barrier()

for i in range(n):
    dj_balanced.measure(i,i)

dj_balanced.draw('mpl')


```

```python
#Making circuit with balanced blackbox

dj_const = QuantumCircuit(n+1, n)

for qubit in range(n):
    dj_const.h(qubit)

dj_const.x(n)
dj_const.h(n)

# Add the blackbox we want
dj_const += const_oracle

for qubit in range(n):
    dj_const.h(qubit)
dj_const.barrier()

for i in range(n):
    dj_const.measure(i,i)

dj_const.draw('mpl')
```

```python
#Running circuits on quantum computer

balanced_job = execute(dj_balanced, backend=qcomp, shots=1024)
job_monitor(balanced_job)

const_job = execute(dj_const, backend=qcomp, shots=1024)
job_monitor(const_job)
```

```python
#Plotting balanced circuits result
balanced_res = balanced_job.result()
plot_histogram(balanced_res.get_counts())
```

```python
#Plotting constant circuits result
const_res = const_job.result()
plot_histogram(const_res.get_counts())
```

```python
#The errors are due to errors in quantum computation. The balanced circuit result had a bit too many errors, but we can see that the most likely values are towards the 111 side.
#The constant circuit results, on the other hand, look clean, and the results are clearly 000.

#We are finally able to beat classical computers in a few tasks!
```
