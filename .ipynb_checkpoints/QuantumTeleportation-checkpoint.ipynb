```python
%matplotlib inline
import qiskit
from qiskit import *
from qiskit import IBMQ
from qiskit.tools.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor

IBMQ.load_account()
sim = Aer.get_backend('qasm_simulator')
provider = IBMQ.get_provider('ibm-q'); qcomp = provider.get_backend('ibmqx2')
vector_sim = Aer.get_backend('statevector_simulator')

print(qiskit.__qiskit_version__)
```

```python
#This is the Quantum Teleportation Algorithm, where the state of one qubit is transferred to another qubit. Note that the first qubit loses its state once it passes on - meaning that you can't clone something
qc = QuantumCircuit(3, 3)

#Encoding qubit state - This could be a superposition as well, but then we couldn't verify easily, so we will use 0 or 1
message = input("Enter state of qubit you want to send")
if message == "0":
    pass
else:
    qc.x(0)
qc.barrier()

#Entangling qubit 2 and 3
qc.h(1)
qc.cx(1, 2)

#Entangling qubit 1 and 2
qc.cx(0, 1)
qc.h(0)

qc.barrier()

#Measuring states of first 2 qubits since last operation is dependent
qc.measure([0, 1], [0, 1])
qc.barrier()

#Finishing operations on last qubit to get final state
qc.cx(1, 2)
qc.cz(0, 2)

#Finally Measuring qubit state to verify
qc.measure(2, 2)

qc.draw('mpl')
```

```python
#Expect to see same state as one entered above. The qubit to look for is the bottom most one since it corresponds to the second qubit. They should be all 0s, or all 1s.
sim_result = execute(qc, backend = sim, shots=1024).result()
plot_histogram(sim_result.get_counts())
```
