```python
%matplotlib inline
import qiskit
from qiskit import *
from qiskit import IBMQ
from qiskit.tools.visualization import plot_histogram
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.tools.monitor import job_monitor

IBMQ.load_account()
sim = Aer.get_backend('qasm_simulator')
provider = IBMQ.get_provider('ibm-q'); qcomp = provider.get_backend('ibmqx2')
vector_sim = Aer.get_backend('statevector_simulator')

print(qiskit.__qiskit_version__)
```

```python


#This circuit implements quantum fourier transforms, which takes in waves and outputs frequency, just like the normal fourier transofrm.
#It is interesting as when you input superposition states, this transform outputs a superposition of the transform of the indiviudal qubits.

from numpy import pi
def qft_rotations(circuit, n):

    if n == 0:
        return circuit
    n -= 1
    circuit.h(n)
    for qubit in range(n):
        circuit.cu1(pi/2**(n-qubit), qubit, n)


    qft_rotations(circuit, n)

#Checking the circuit
qc = QuantumCircuit(4)
qft_rotations(qc,4)
qc.draw('mpl')
```

```python
#We also need to swap qubits after the transform in the manner shown in the picture

def swap_registers(circuit, n):
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    return circuit

def qft(circuit, n):
    """QFT on the first n qubits in circuit"""
    qft_rotations(circuit, n)
    swap_registers(circuit, n)
    return circuit

#Seeing how they must be swapped
qc = QuantumCircuit(4)
qft(qc,4)
qc.draw('mpl')
```

```python
#We will be transforming the number 5. In binary, 5 is 0b101, so we need 3 qubits
qc = QuantumCircuit(3)

# Encode the state 5
qc.x(0)
qc.x(2)

qc.draw('mpl')
```

```python
#Let's plot the states of these vectors. Right now, it should just show the number 5, nothing special
statevector = execute(qc, backend=vector_sim).result().get_statevector()
plot_bloch_multivector(statevector)
```

```python
#Adding the fourier transform code to the circuit
qft(qc,3)
qc.draw('mpl')
```

```python
#Awesome, now we will run the cirucit on the simulator, and see the vector values
statevector = execute(qc, backend=vector_sim).result().get_statevector()
plot_bloch_multivector(statevector)
```
